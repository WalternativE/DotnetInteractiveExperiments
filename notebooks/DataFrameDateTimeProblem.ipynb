```
open System
open Microsoft.Data.Analysis
open Microsoft.ML
open System.Collections.Generic
```

```
let numericIdxColumn = PrimitiveDataFrameColumn("NumId", [ 1 .. 300])
let sineColumn = PrimitiveDataFrameColumn("Sine", [ for x in 1. .. 300. do Math.Sin(x / 10.) ])
let cosColumn = PrimitiveDataFrameColumn("Cosine", [ for x in 1. .. 300. do Math.Cos(x / 10.) ])

let cols : DataFrameColumn array = [| numericIdxColumn; sineColumn; cosColumn |]

let frame = DataFrame cols
frame
```

```
// No DateTime column so it works
frame.Explore()
```

```
// No DateTime column so it works
frame.ExploreWithSandDance()
```

```
let dateColumn =
    PrimitiveDataFrameColumn("Date",
                             [ for x in 1. .. 300. do
                                let startDate = DateTime.Today.AddYears(-1)
                                startDate.AddDays(x)])

let cols' : DataFrameColumn array = [| dateColumn; sineColumn; cosColumn |]

let frame' = DataFrame cols'
frame'
```

```
// DateTime column in frame so it does not work
frame'.Explore()
```
